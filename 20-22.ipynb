```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\harsh\Desktop\ML\department_salary.csv"
df = pd.read_csv(file_path)
```

```python
# Get unique departments
departments = df['Department'].unique()

# Count the number of employees in each department
employee_counts = [len(df[df['Department'] == dept]) for dept in departments]

# Plotting the bar chart
plt.bar(departments, employee_counts, color='red')

# Adding labels and title
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees in Each Department')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# Get unique departments
departments = df['Department'].unique()

# Plot salary trends for each department
for dept in departments:
    dept_data = df[df['Department'] == dept]
    plt.plot(dept_data['Name'], dept_data['Salary'], marker='o', label=dept)

# Adding labels and title
plt.xlabel('Employee Name')
plt.ylabel('Salary')
plt.title('Salary Trend by Department')
plt.legend(title="Department")
plt.grid(axis='y', linestyle='--')

# Display the plot
plt.show()
```
